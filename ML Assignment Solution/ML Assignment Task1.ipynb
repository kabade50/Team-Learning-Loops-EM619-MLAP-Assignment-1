```python
import os
import pandas as pd # type: ignore
import numpy as np # type: ignore
import matplotlib.pyplot as plt # type: ignore
import seaborn as sns # type: ignore
from sklearn.decomposition import PCA # type: ignore
from sklearn.preprocessing import StandardScaler # type: ignore
import tsfel  # Time Series Feature Extraction Library # type: ignore

# Enable interactive mode for Matplotlib
plt.ion()

# Define root directory
root_directory = r"D:\MLFolder\UCI HAR Dataset\UCI HAR Dataset\Combined"

# List to store file paths and activity labels
subject_files = ["D:\MLFolder\\UCI HAR Dataset\\UCI HAR Dataset\\Combined\\Train\\LAYING\\Subject_6.csv",
                 "D:\MLFolder\\UCI HAR Dataset\\UCI HAR Dataset\\Combined\\Train\\SITTING\\Subject_6.csv",
                 "D:\MLFolder\\UCI HAR Dataset\\UCI HAR Dataset\\Combined\\Train\\STANDING\\Subject_6.csv",
                 "D:\MLFolder\\UCI HAR Dataset\\UCI HAR Dataset\\Combined\\Train\\WALKING\\Subject_6.csv",
                 "D:\MLFolder\\UCI HAR Dataset\\UCI HAR Dataset\\Combined\\Train\\WALKING_DOWNSTAIRS\\Subject_6.csv",
                 "D:\MLFolder\\UCI HAR Dataset\\UCI HAR Dataset\\Combined\\Train\\WALKING_UPSTAIRS\\Subject_6.csv"
]
activity_labels = ["Laying","Sitting","Standing","Walking","Walking_Downstairs","Walking_upstairs"]

# Recursively find all subject_6.csv files
for subdir, _, files in os.walk(root_directory):
    for file in files:
        if file == "subject_6.csv":
            file_path = os.path.join(subdir, file)
            activity_name = os.path.basename(subdir)  # Extract activity name
            subject_files.append(file_path)
            activity_labels.append(activity_name)

# Check if any files are found
if not subject_files:
    print("No subject_6.csv files found!")
    exit()

# Function to perform PCA and visualize
def perform_pca_and_plot(data, title):
    try:
        num_samples, num_features = data.shape
        if num_samples < 2 or num_features < 2:
            print(f"Skipping PCA for {title} (Not enough samples/features)")
            return
        
        # Standardize Data
        scaler = StandardScaler()
        scaled_data = scaler.fit_transform(data)

        # Apply PCA
        n_components = min(2, num_features, num_samples)
        pca = PCA(n_components=n_components)
        pca_result = pca.fit_transform(scaled_data)

        # Plot PCA Results
        plt.figure(figsize=(8, 6))
        sns.scatterplot(x=pca_result[:, 0], y=pca_result[:, 1] if n_components > 1 else np.zeros_like(pca_result[:, 0]), alpha=0.6)
        plt.title(f"PCA Visualization: {title}")
        plt.xlabel("Principal Component 1")
        plt.ylabel("Principal Component 2" if n_components > 1 else "Zero Axis (Only 1 Component)")
        plt.grid()
        plt.show(block=True)
        plt.pause(1)

    except Exception as e:
        print(f"PCA Failed for {title}: {e}")

        # Check if we can use pairplot (at least 2 features)
        df_fallback = pd.DataFrame(scaled_data, columns=data.columns)
        df_fallback = df_fallback.dropna(axis=1)  # Remove NaN columns

        if df_fallback.shape[1] >= 2:
            print(f"falling back to pairplot for {title}...")
            sns.pairplot(df_fallback)
            plt.show(block=True)
            plt.pause(1)
        else:
            print(f"Showing raw scatterplot")

            plt.figure(figsize=(8, 6))
            plt.scatter(range(len(df_fallback)), df_fallback.iloc[:, 0], alpha=0.6)
            plt.title(f"Scatter Plot: {title}")
            plt.xlabel("Index")
            plt.ylabel(df_fallback.columns[0])
            plt.grid()
            plt.show(block=True)
            plt.pause(1)

for file_path, activity_name in zip(subject_files, activity_labels):
    print(f"\nProcessing file: {file_path} ({activity_name})")

    try:
        # Read CSV
        data = pd.read_csv(file_path)

        # Validate DataFrame
        if data.empty:
            print(f"Skipping {file_path} (Empty file)")
            continue

        # Standardize column names
        data.columns = data.columns.str.strip().str.lower()

        # Required columns
        required_cols = {"accx", "accy", "accz"}
        if not required_cols.issubset(set(data.columns)):
            print(f"Skipping {file_path} (Missing columns: {required_cols - set(data.columns)})")
            continue

        # Compute total acceleration squared
        data["total_acc_sq"] = data["accx"]**2 + data["accy"]**2 + data["accz"]**2

        # Perform PCA on Total Acceleration
        perform_pca_and_plot(data[["total_acc_sq"]], f"Total Acceleration ({activity_name})")

        # Step 2: PCA on TSFEL Features
        cfg = tsfel.get_features_by_domain("all")  # Extract all features
        tsfel_features = tsfel.time_series_features_extractor(cfg, data[["accx", "accy", "accz"]], verbose=0)

        # Drop NaN values
        tsfel_features = tsfel_features.dropna(axis=1)

        # Perform PCA on TSFEL Features
        perform_pca_and_plot(tsfel_features, f"TSFEL Features ({activity_name})")

        # Step 3: PCA on Dataset Features
        dataset_features = data.drop(columns=["total_acc_sq"], errors="ignore")  # Remove computed feature

        # Perform PCA on Dataset Features
        perform_pca_and_plot(dataset_features, f"Dataset Features ({activity_name})")

    except Exception as e:
        print(f"Error processing {file_path}: {e}")

```

```python

import os
import pandas as pd  # type: ignore
import numpy as np  # type: ignore
import matplotlib.pyplot as plt  # type: ignore
import seaborn as sns  # type: ignore
from sklearn.decomposition import PCA  # type: ignore
from sklearn.preprocessing import StandardScaler  # type: ignore
import tsfel  # Time Series Feature Extraction Library  # type: ignore

# Enable interactive mode for Matplotlib
plt.ion()

# File paths and activity labels
subject_files = [
    "D:\\MLFolder\\UCI HAR Dataset\\UCI HAR Dataset\\Combined\\Train\\LAYING\\Subject_6.csv",
    "D:\\MLFolder\\UCI HAR Dataset\\UCI HAR Dataset\\Combined\\Train\\SITTING\\Subject_6.csv",
    "D:\\MLFolder\\UCI HAR Dataset\\UCI HAR Dataset\\Combined\\Train\\STANDING\\Subject_6.csv",
    "D:\\MLFolder\\UCI HAR Dataset\\UCI HAR Dataset\\Combined\\Train\\WALKING\\Subject_6.csv",
    "D:\\MLFolder\\UCI HAR Dataset\\UCI HAR Dataset\\Combined\\Train\\WALKING_DOWNSTAIRS\\Subject_6.csv",
    "D:\\MLFolder\\UCI HAR Dataset\\UCI HAR Dataset\\Combined\\Train\\WALKING_UPSTAIRS\\Subject_6.csv"
]
activity_labels = ["Laying", "Sitting", "Standing", "Walking", "Walking_Downstairs", "Walking_Upstairs"]

def extract_tsfel_features(data):
    """Extracts TSFEL features from accelerometer data."""
    try:
        cfg = tsfel.get_features_by_domain("all")  # Extract all features
        tsfel_features = tsfel.time_series_features_extractor(cfg, data, verbose=0)
        return tsfel_features.dropna(axis=1)  # Drop NaN values
    except Exception as e:
        print(f"Error extracting TSFEL features: {e}")
        return pd.DataFrame()

def perform_pca_and_plot(data, title, labels=None):
    """Performs PCA and visualizes the results."""
    try:
        if data.shape[1] < 2:
            print(f"Skipping PCA for {title} (Not enough features)")
            return
        
        # Standardize Data
        scaler = StandardScaler()
        scaled_data = scaler.fit_transform(data)

        # Apply PCA
        pca = PCA(n_components=2)
        pca_result = pca.fit_transform(scaled_data)

        # Plot PCA Results
        plt.figure(figsize=(8, 6))
        sns.scatterplot(x=pca_result[:, 0], y=pca_result[:, 1], hue=labels, alpha=0.6)
        plt.title(f"PCA Visualization: {title}")
        plt.xlabel("Principal Component 1")
        plt.ylabel("Principal Component 2")
        plt.grid()
        plt.legend()
        plt.show(block=True)
        plt.pause(1)
    except Exception as e:
        print(f"PCA Failed for {title}: {e}")

def process_all_files(file_paths, activity_names):
    """Processes multiple CSV files and creates a combined PCA scatter plot."""
    all_pca_data = []
    all_labels = []
    tsfel_pca_data = []
    tsfel_labels = []
    
    for file_path, activity_name in zip(file_paths, activity_names):
        try:
            # Read CSV
            data = pd.read_csv(file_path)
            
            if data.empty:
                print(f"Skipping {file_path} (Empty file)")
                continue
            
            # Standardize column names
            data.columns = data.columns.str.strip().str.lower()

            required_cols = {"accx", "accy", "accz"}
            if not required_cols.issubset(set(data.columns)):
                print(f"Skipping {file_path} (Missing columns: {required_cols - set(data.columns)})")
                continue

            # Compute total acceleration squared
            data["total_acc_sq"] = data["accx"]**2 + data["accy"]**2 + data["accz"]**2

            # Perform PCA on acceleration components
            scaler = StandardScaler()
            scaled_acc = scaler.fit_transform(data[["accx", "accy", "accz", "total_acc_sq"]])
            pca = PCA(n_components=2)
            pca_acc = pca.fit_transform(scaled_acc)
            
            all_pca_data.append(pca_acc)
            all_labels.extend([activity_name] * len(pca_acc))

            # Extract TSFEL features and apply PCA
            tsfel_features = extract_tsfel_features(data[["accx", "accy", "accz"]])
            if not tsfel_features.empty:
                scaler_tsfel = StandardScaler()
                scaled_tsfel = scaler_tsfel.fit_transform(tsfel_features)
                if scaled_tsfel.shape[1] >= 2:
                    pca_tsfel = PCA(n_components=2).fit_transform(scaled_tsfel)
                    tsfel_pca_data.append(pca_tsfel)
                    tsfel_labels.extend([activity_name] * len(pca_tsfel))
        
        except Exception as e:
            print(f"Error processing {file_path}: {e}")
    
    if all_pca_data:
        all_pca_data = np.vstack(all_pca_data)
        perform_pca_and_plot(pd.DataFrame(all_pca_data, columns=["PC1", "PC2"]), "PCA on Total Acceleration (All Activities)", labels=all_labels)
    
    if tsfel_pca_data:
        tsfel_pca_data = np.vstack(tsfel_pca_data)
        perform_pca_and_plot(pd.DataFrame(tsfel_pca_data, columns=["PC1", "PC2"]), "PCA on TSFEL Features (All Activities)", labels=tsfel_labels)

# Process the files
process_all_files(subject_files, activity_labels)
```

```python
import numpy as np # type: ignore
import pandas as pd # type: ignore
import seaborn as sns # type: ignore
import matplotlib.pyplot as plt # type: ignore
import tsfel # type: ignore
from sklearn.preprocessing import StandardScaler # type: ignore
import os

# Define dataset path
data_file = "D:\\MLFolder\\UCI HAR Dataset\\UCI HAR Dataset\\Combined\\Train\\LAYING\\Subject_6.csv"

# Ensure file exists
if not os.path.exists(data_file):
    raise FileNotFoundError(f"Data file not found: {data_file}")

# Load dataset
df = pd.read_csv(data_file)

# Ensure column names are clean
df.columns = df.columns.str.strip().str.lower()

# Assign activity manually (since 'y_train.txt' is missing)
df['activity'] = "LAYING"

# Select original dataset features (acceleration values)
original_features = df[['accx', 'accy', 'accz']]

# Extract TSFEL features
cfg = tsfel.get_features_by_domain()
window_size = 100  # Adjust window size based on dataset
tsfel_features = tsfel.time_series_features_extractor(cfg, df[['accx', 'accy', 'accz']], window_size=window_size, verbose=0)

# Step 1: Combine TSFEL Features and Dataset Features
combined_features = pd.concat([tsfel_features, original_features], axis=1)

# Step 2: Compute the Correlation Matrix
correlation_matrix = combined_features.corr()

# Step 3: Identify Highly Correlated Features (Threshold > 0.9)
high_correlation_pairs = []
threshold = 0.9  # Can be adjusted

for i in range(len(correlation_matrix.columns)):
    for j in range(i + 1, len(correlation_matrix.columns)):  # Avoid self-correlation
        if abs(correlation_matrix.iloc[i, j]) > threshold:
            feature1 = correlation_matrix.columns[i]
            feature2 = correlation_matrix.columns[j]
            corr_value = correlation_matrix.iloc[i, j]
            high_correlation_pairs.append((feature1, feature2, corr_value))

# Step 4: Display the Highly Correlated Features
if high_correlation_pairs:
    print("\n🔍 Highly Correlated Features (|correlation| > 0.9):")
    for feat1, feat2, corr in high_correlation_pairs:
        print(f"📌 {feat1} ↔ {feat2} (Correlation: {corr:.2f})")
else:
    print("✅ No highly correlated features found.")

# Step 5: Heatmap for Visualization
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, cmap="coolwarm", annot=False, fmt=".2f")
plt.title("Feature Correlation Matrix")
plt.show()
```
